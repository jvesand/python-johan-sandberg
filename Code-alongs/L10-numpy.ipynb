```python
numbers_list = [1,2,3]
numbers_list = numbers_list*2
print(numbers_list)

# in a list we can place different data types
mixed_list = [1, 'ett', True, 2.3]
# dynamic typing in Python (most languages static, and require typing like int x = 3, not just x)
# each element in the list is an object, it has information on what type it is
# this requires some overhead, which slows down the program
# note that when we have one data type, it is unncessary to have this overhead
print(mixed_list)
```

```python
# in numpy we have statically typed
import numpy as np

# create an array from a list
vector1 = np.array([2,5,1])
print(vector1)

# element wise multiplication
vector2 = vector1*2
print(vector2)

# element wise addition
print(f"{vector1+2=}")

# indexing
print(f"{vector1[2]=}")
print(f"{vector1[-1]=}")

# slicing
print(f"{vector1[1:]=}")

# tilldela new value
vector1[1] = 1337
print(vector1)

# can't append to an array (can't change its size)
# can't have different types in an array

# sort
vector1.sort()
print(f"{vector1=}")

# sum
print(f"{vector1.sum()=}")
print(f"{vector1.mean()=:.2f}")


# numpy faster than builtin because of vectorization and static typing
# it also pushes looping to compilation layer (in C): should not loop numnpy??
# https://realpython.com/numpy-array-programming/
```

```python
ones_vector = np.ones(10)
print(f"{ones_vector=}")

# 2d array of shape (3,3)
matrix_3x3 = np.ones((3,3))

print(f"{matrix_3x3=}")

# indexing [row, column] - counts from 0
matrix_3x3[2,1] = 10

matrix_3x3
```

```python
np.zeros((2,5)) # shapeLike - takes in a tuple (2,5) or a list [2,5]
```

```python
print(matrix_3x3.shape)
print(matrix_3x3)
# instead of looping through lines:
# use slicing operator to get a row
print(f"{matrix_3x3[2,:]=}") # colon is all (columns in this case)

# slicing operator to get a column
print(f"{matrix_3x3[:,1]=}") # colon is all (rows in this case)

# add 2 vectors
matrix_3x3[2,:] + matrix_3x3[:,1]
```

```python
matrix_3x3[2,0] = 12
matrix_3x3[1,2] = 5
print(matrix_3x3)

# same
print(f"{matrix_3x3[2,:]=}")
print(f"{matrix_3x3[2]=}")

# same, all rows but last one
print(f"{matrix_3x3[:2]=}")
print(f"{matrix_3x3[:-1]=}")

# slicing to get column
print(f"{matrix_3x3[:,2]=}")
```

## Performance

```python
%%timeit
import random as rnd

number_dices = 1_000_000

many_dices = []

for _ in range(number_dices):
    many_dices.append(rnd.randint(1,6))
```
