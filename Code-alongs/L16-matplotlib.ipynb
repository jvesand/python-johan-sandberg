```python
import matplotlib.pyplot as plt # plt in namespace (module in sys.modules)
import numpy as np # np in namespace (module in sys.modules)

# array from -10 to 10 with 100 points equally spaced
x = np.linspace(-10, 10, 100)
x = np.linspace(-2*np.pi, 2*np.pi, 100)

# difference f = np.cos(x) not a function, uses x variable
f = lambda x: np.cos(x)
g = lambda x: np.sin(x)

f(x)
```

## MATLAB style interface

```python
plt.figure(figsize=(8,4), dpi=200)
plt.plot(x, f(x))
plt.plot(x, g(x), color="red")

plt.title("cos(x), sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks([-2*np.pi, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 2*np.pi])
```

## OOP approach

```python
# instantiated fig, ax from figure and axes
fig, ax = plt.figure(figsize = (8,4)), plt.axes()

ax.plot(x, f(x)) # ax object has a plot method
ax.set(xlabel = "x", ylabel = "y", title = "$\cos{x}$", xlim = (-5, 5), ylim = (-1.2, 1.2))

fig.savefig("../Data/cos(x).png", dpi=200)
```

## Subplots

```python
# MATLAB style
plt.figure(figsize=(16,5))
plt.subplot(1,2,1) # (rows, columns, panel number)
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.title("cos(x) and sin(x)")


plt.subplot(1,2,2)
plt.plot(x, f(x))
plt.title("cos(x)")

plt.suptitle("Trigonometric functions", fontweight = "bold")
```

```python
## OOP approach
fig, ax = plt.subplots(1,2, figsize=(16,5))
ax[0].plot(x, f(x))
ax[0].plot(x, g(x))
ax[1].plot(x, g(x))

ax[0].set(title = "cos(x) & sin(x)")
fig.suptitle("Big title", fontsize = 25, fontweight = "bold")

# if we would have 2x2 -> ex 3rd ax: ax[1][0]
```

## Example plots

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# randn: normal distribution (standard mean 0 sd 1)
data_normal = np.random.randn(1000)

ax[0].hist(data_normal, bins=30, alpha=0.7)
ax[0].set(title="Histogram of normally distributed data", xlabel="x", ylabel="y")

# like doing plt.scatter (rather plt.subplots[1].scatter)
# format with alt-shift-f
scatter = ax[1].scatter(
    x=data_normal,
    y=np.random.randn(1000),
    s=100*np.random.randn(1000),
    alpha=0.7,
    cmap="viridis",
    c=np.random.randn(1000),
)

ax[1].set(title="Bubble chart", xlabel="x", ylabel="y")
fig.colorbar(scatter)
fig.set_facecolor("lightblue")
```
