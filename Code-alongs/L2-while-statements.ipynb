# while loop

- "if-loop"
- as long as a condition is True -> the statements inside the while loop will run

```python
i = 0

while i < 4:
    print(f"Hej {i=}")
    i += 1 # adds one to i <=> i = i +1
print("Program continues here")
```

## Example - birds and oil

Oil leakage: 1/2 each year. Start with 80000 birds, how long for 1/10

```python
birds = 8e4
years = 0

# alternative
birds, years = 8e4, 0

while birds > 8e4/10:
    birds /= 2
    years += 1
    print(f"Year {years}: {birds:.0f} birds left")
print(f"It takes {years} years")
```

# Example - Milk

1 liter of milk has 1.5e6 bacterias
increaseds with 50% each hour in room temp
Gets soured after 1e7 bacterias
Hours until milk sours

```python
import matplotlib.pyplot as plt

bacterias = 1.5e6
hours = 0

bacterias_list = [bacterias]
print(bacterias_list)

while bacterias < 1e7:
    bacterias *= 1.5
    bacterias_list.append(bacterias)
    hours += 1

print(bacterias_list)
print(f"It takes {hours} hours for the milk to sour")

# creates a range object with numbers spaced 1
hours_list = range(0, hours+1)

plt.plot(hours_list, bacterias_list, 'o-')
plt.plot(hours_list, [1e7]*(hours+1))
print(f"{[1e7]*(hours+1)}")
plt.grid()
plt.xlabel("Number of hours")
plt.ylabel("Number of bacterias")
plt.title("Bacterias in milk in room temperature")
```
