# Functions

```python
# define a function with 1 parameter
def cuber(x):
    return x**3

# calls the function cuber() using 1 argument
print(f"{cuber(5)=}")

cubes = [cuber(i) for i in range(10)]
cubes
```

```python
import matplotlib.pyplot as plt

# semicolon mutes last output
plt.plot(cubes);
```

```python
# 3 parameters
def my_mean(number1, number2, number3):
    print(f"{number1=}, {number2=}, {number3=}")
    return (number1 + number2 + number3)/3

print(my_mean(1,2,3))

double_mean = my_mean(1,2,3) * 2
print(double_mean)
```

## Default values

```python
# void function (no return value)
def hello(number_of_greetings, name = "buddy", age = 0):
    if age < 0:
        print("Age has to be >= 0")
        return
    for i in range(number_of_greetings):
        print(f"Hej {name}, you are {age} years old")

hello(2)

hello(1, "Bella", 1)

hello(1, "Bella", -1)

# positional (number_of_greetings=1) and keyword arguments (age=)
hello(1, age=34)
```

```python
def word_counter(text = ""):
    word_list = text.split()
    return len(word_list)

quote = "No I am your father - Luke Skywalker"
word_counter(quote)
```

# Arbitrary arguments - *args

```python
# arbritrary number of arguments
def my_mean2(*numbers):
    sum_ = 0 # to not overwrite sum function
    for number in numbers:
        sum_ += number
    return sum_/len(numbers)

my_mean2(1,2,3,4,5)
```

## Draw lines

```python
# x could be x_list
def draw_lines(x, k=1, m=1, title = ""):
    y = [k*x_+m for x_ in x]

    plt.plot(x,y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(title)

x = list(range(-10,10))
draw_lines(x, title = "Lines")
draw_lines(x, k=-2, m=-1, title = "test")

# find intersection point and plot it
# line1 = x+1
# line2 = -2x -1
# 3x = -2 -> x = -2/3
# intersection = [-2/3, (-2/3)]
plt.plot(-2/3, (-2/3)+1, 'o')

```
