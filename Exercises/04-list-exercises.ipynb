# 5. Dice rolls convergence

```python
import random as rnd

rolls = [rnd.randint(1,6) for _ in range(100)]
sixes = [i for i in rolls if i == 6]

print(f"{sixes=}")

# test with lambda function
lambda_filtered = filter(lambda roll: roll == 6, rolls)

print(f"{lambda_filtered=}")
print(f"{[a for a in lambda_filtered]=}")
print(list(lambda_filtered))

# i have overwritten list function with variable!
# check with str() if its a function
str(list)

# equivalent to the lambda function
def filter_sixes(roll):
    if (roll == 6):
        return True
    else:
        return False

function_filtered = filter(filter_sixes, rolls)

for a in function_filtered:
    print(a)

print(f"{function_filtered=}")
```

```python
import random as rnd
import matplotlib.pylab as plt
rnd.seed(1)

rolls_list, prob_list = [], []
num_rolls = [10**i for i in range(1,7)]

for i in num_rolls:
    rolls = [rnd.randint(1,6) for _ in range(i)]
    rolls_list.append(rolls)

for i in rolls_list:
    sixes = [j for j in i if j == 6]
    prob_list.append(len(sixes)/len(i))

print(f"{prob_list=}")
plt.plot(prob_list, '-*')
plt.xticks([0,1,2,3,4,5], num_rolls)
plt.xlabel("Number of dice rolls")
plt.ylabel("Probability of six")
plt.title("Dice rolls convergence");
```

# 6. Monte Carlo simulation

```python
import random as rnd
import math
import matplotlib.pylab as plt

x_list = [rnd.uniform(-1,1) for _ in range(5000)]
y_list = [rnd.uniform(-1,1) for _ in range(5000)]
outer_points, inner_points = [], []

for i in range(len(x_list)):
    distance = math.sqrt(x_list[i]**2 + y_list[i]**2)
    if distance > 1:
        # plt.plot(x_list[i], y_list[i], "o", color='green')
        outer_points.append([x_list[i], y_list[i]])
    else:
        # plt.plot(x_list[i], y_list[i], "o", color='blue')
        inner_points.append([x_list[i], y_list[i]])

# only way to use the list of tuples?
plt.plot([i[0] for i in outer_points], [i[1] for i in outer_points], "o", color = 'green')
plt.plot([i[0] for i in inner_points], [i[1] for i in inner_points], "o", color = 'blue')

# area circle / area square 3,14/4 = 0,785
len(inner_points)/len(outer_points + inner_points)
```

# 7. A cute rabbit among two ferocious snakes

```python
# you get it right 2/3 by switching, rather than 1/3 by staying

import random as rnd
import matplotlib.pylab as plt

num_reps = [10**i for i in range(1,7)]

# rabbit is in 1
def staying(reps):
    wins = 0
    for i in range(reps):
        choice = rnd.randint(1,3)
        if choice == 1:
            wins += 1
    return wins/reps

def switching(reps):
    wins = 0
    for i in range(reps):
        choice = rnd.randint(1,3)
        if choice != 1:
            wins += 1
    return wins/reps

staying_list = [staying(i) for i in num_reps]
switching_list = [switching(i) for i in num_reps]

plt.plot(staying_list, "o-", color="blue")
plt.plot(switching_list, "*-", color="green")
plt.xticks([0,1,2,3,4,5], num_reps)
plt.xlabel("Number of reps")
plt.ylabel("Win probability")
plt.title("Rabbits and snakes")
```
