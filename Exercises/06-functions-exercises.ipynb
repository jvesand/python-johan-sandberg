# 2. Euclidean distance

```python
import math

point1 = input("Input the first point: (p1, p2)")
point2 = input("Input the second point: (p1, p2)")

point1 = [int(i) for i in point1.split(",") if i.isdigit()]
point2 = [int(i) for i in point2.split(",") if i.isdigit()]

def euclidean_distance(point1, point2):
    return math.sqrt((point1[0] - point2[0])**2 + \
        (point1[1] - point2[1])**2)

euclidean_distance(point1, point2)
```

```python
points = [[10,3], [-1,-9], [10,-10], [4,-2], [9,-10]]

for i in points:
    print(f"The distance between [0, 0] and {i} is {euclidean_distance([0,0], i):.1f}")
```

# 3. Mathematical functions

```python
import matplotlib.pylab as plt
import numpy as np

fx = lambda x: x**2 -3
gx = lambda x: 4*x - 7

x = np.linspace(-10, 10, 100)

plt.plot(x, fx(x), "-")
plt.plot(x, gx(x), "-")
plt.title("Functions")
plt.xlabel("x")
plt.ylabel("y")
```

# 4. Name cleaner

```python
string_list = ["  MaRcUs ", " iDA aNderSon", "OLOF Olofsson            "  ]

def name_cleaner(string_list):
    string_list = [i.strip().lower().title() for i in string_list]
    print("\n".join(string_list))

name_cleaner(string_list)
```

# 5. Change

```python
import re

bank_notes = ["1000-lapp", "500-lapp", "200-lapp", "100-lapp", \
    "50-lapp", "20-lapp", "10-krona", "5-krona", "2-krona", "1-krona"]
bank_digits = [int(re.sub("\D", "", i)) for i in bank_notes]

def change(moneys):
    print(f"Breakdown of {moneys}:\n")
    for i in range(len(bank_digits)):
        if moneys/bank_digits[i] >= 1:
            print(f"{int(moneys/bank_digits[i])} {bank_notes[i]}")
            moneys = moneys % bank_digits[i]

user_input = int(input("Input money: "))
change(user_input)
```
