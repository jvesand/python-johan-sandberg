# 2. Several dices

```python
# np.meshgrid(), np.add.reduce(), np.unique()

# np.sum(a) returns sum of all elements in the matrix
# np.add.reduce sums up each column and returns a vector of sums
# np.add.reduce(a) will sum along axis=0 so is equivalent to sum(a, axis=0)

dice = [1,2,3,4,5,6]
# dice = np.linspace(1, 6, 6)

# np.meshgrid makes ordered (x,y) pairs from (1,1) to (6,6)
aa, bb = np.meshgrid(dice, dice)

print(f"{aa=}")
print(f"{bb=}")

print(f"all possible sums:\n{aa+bb=}")
print(f"only unique sums:\n{np.unique(aa+bb)=}")

# count number of each value
occurences = [np.count_nonzero(aa+bb == i) for i in np.unique(aa+bb)]
print(f"{occurences=}")

# count frequences
frequencies = [i/36 for i in occurences]
print(frequencies)
print(np.sum(frequencies))

# plot probability distribution of sums
import matplotlib.pyplot as plt
plt.bar(np.unique(aa+bb), frequencies)
plt.xlabel("Unique occurences")
plt.ylabel("Probablility")
plt.title("Two dice sum probability distribution")
```

```python
threes = [aa+bb+np.ones((6,6))*i for i in range(1,7)]

frequences_three = []

for i in range(len(np.unique(threes))):
    count = [np.count_nonzero(j == np.unique(threes)[i]) for j in threes]
    frequences_three.append(np.sum(count))

frequences_three = [i/6**3 for i in frequences_three]

plt.bar(np.unique(threes), frequences_three)
```
